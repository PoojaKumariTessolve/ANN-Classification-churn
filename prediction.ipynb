```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler

import pickle
import pandas as pd
import numpy as np
```

```python
# load the ann train model , scaler pickel file, one hot encodeing

model = load_model('model.h5')

# load the encoader and scaler

with open('label_encoder_geo.pkl', 'rb') as file:
    label_enoder_geo = pickle.load(file)

with open('label_encoder_gender.pkl', 'rb') as file:
    label_enoder_gender = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    label_enoder_scaler = pickle.load(file)
```

```python
import os
os.getcwd()
os.listdir()
```

```python
input_data = {
    'CreditScore' : 600,
    'Geography' : 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure' : 3,
    'Balance' : 6000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 5000
}
```

```python
geo_encoded = label_enoder_geo.transform(
    [[input_data['Geography']]]
).toarray()

geo_encoded_df = pd.DataFrame(
    geo_encoded,
    columns=label_enoder_geo.get_feature_names_out(['Geography'])
)
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender']=label_enoder_gender.transform(input_df['Gender'])
input_df
```

```python
input_df = pd.concat([input_df.drop("Geography" , axis= 1), geo_encoded_df], axis=1)
input_df
```

```python
# scaling the input data
input_scaled = label_enoder_scaler.transform(input_df)
input_scaled
```

```python
## prediction churn
prediction =model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
prediction_proba
```

```python
if prediction_proba > 0.5:
    print('The customer is likely to churn')
else :
    print('The customer is not likely to churn')
```
